### Part I: Instructions

In this part of the assignment, your task is to create a one-dimensional random walk model similar to the one in [the example notebook](./mod01-PythonIntro-1.ipynb) with the following parameters:

* `Nt` = 200
* `mu_x` = 1.5
* `s_x` = 2.1
* `min_x` = 1.25

Where `min_x` is a new parameter that is the _minimum step size_ at each step in the random walk. When you calculate the step size at each step, you will only accept values that are larger than `min_x` by using a `while` loop to keep generating step sizes until the step size exceeds `min_x`. Do this for `Nx` = 1,000 random walkers. 

___Deliverables:___ 

1. Code to create the random walk with comments
2. Reproduce the 2 figures in section 5.1 and 5.2 of the example notebook (__Note__: you will have to adjust the `bins` and `range` parameters provided to histogram). 

```python
import numpy as np
import matplotlib.pyplot as plt

Nt = 200
Nx = 1000

mu_x = 1.5
s_x = 2.1
min_x = 1.25
```

### 1. Code to create the random walk

```python
# Create a container for all walkers at every time step. The array will
# will be a container of Nt rows and Nx colmns. Each column corresponds to a
# specific walker and rows corresponds to specific time.
X = np.zeros((Nt,Nx))

# We loop over each time step using a for loop
for i in np.arange(Nt):
    
    # loop over each walker using a nested for loop
    for j in np.arange(Nx):
        
        # initial step is 0
        x_i = 0
        
        # Calculate the random step size for the j^th walker at the i^th time step
        # Computation is repeated until the random step size is greater than minimum step size min_x
        while (x_i<min_x):
            x_i = s_x * np.random.randn() + mu_x
            
        # Check if at initial time step
        if(i==0):
            # At the first step, walker is at rest
            X[i,j] = 0.0 + x_i
        else:
            # The subsequent step follows from the prevous step
            X[i,j] = X[i-1,j] + x_i
        
```

### 2.1  Plot 1: Distance, Mean distance, confidence band

```python
# Calculate the average distance of all walkers through time
mean_X = np.mean(X,axis=1)
# Calculate the standard deviation of all walkers through time
std_X = np.std(X,axis=1)

# The plot in 5.1
plt.figure(figsize=(16,10))

plt.plot(np.arange(Nt), X, 'b-')
plt.plot(np.arange(Nt), mean_X, 'k-')
plt.plot(np.arange(Nt), mean_X + std_X, 'k--')
plt.plot(np.arange(Nt), mean_X - std_X, 'k--')
plt.xlabel('Number of Steps')
plt.ylabel('Distance Traveled [m]')
plt.show()
```

### Part II: Instructions

In this part of the assignment, your task is to create a __two-dimensional__ random walk model. The 2-D random walk is just an extension of the 1-D random walk into a second dimension. Assume that at each step in the random walk, the step size in the x- and y-directions are _independent_ of one another. That is, you do not need to know anything about the step size in the x-direction to calculate the step size in the y-direction and vice versa. Use the following parameters:

* Nt = 300
* Nx = 3
* mu_x = 0.5
* s_x = 1.5
* mu_y = 5.0
* s_y = 2.5

Where `mu_y` and `s_y` are, respectively, the average and standard deviation of the step size in the y-direction.

___Deliverables___

1. Code to create the random walk with comments
2. Plot the position of all three random walkers at each step of their respective random walks on the ___same plot___.

```python
Nt = 300
Nx = 3

mu_x = 0.5
s_x = 1.5
mu_y = 5.0
s_y = 2.5
```

 ### 1. Code to create the random walk


```python
# Create a containers for all walkers at every time step in each direction since the steps in each direction are independent.
#  The array will be a container of Nt rows and Nx colmns. Each column corresponds to a
# specific walker and rows corresponds to specific time.
X = np.zeros((Nt,Nx)); Y = np.zeros((Nt,Nx));

# We loop over each time step using a for loop
for i in np.arange(Nt):
    
    # loop over each walker using a nested for loop
    for j in np.arange(Nx):
        
        
        # Calculate the random step size for the j^th walker at the i^th time step in both directions
        
        x_i = s_x * np.random.randn() + mu_x
        y_i = s_y * np.random.randn() + mu_y
            
        # Check if at initial time step
        if(i==0):
            # At the first step, walker is at (0,0)
            X[i,j] = 0.0 + x_i
            Y[i,j] = 0.0 + y_i
        else:
            # The subsequent step follows from the prevous step
            X[i,j] = X[i-1,j] + x_i
            Y[i,j] = Y[i-1,j] + y_i
        
```

### 2. Plot the position of all three random walkers at each step of their respective random walks

```python
plt.figure(figsize=(16,10))

plt.plot( X[:,0],Y[:,0] ,'b-.',label= 'First walker')
plt.plot(X[:,1], Y[:,1], 'k-.',label='Second walker')
plt.plot(X[:,2], Y[:,2], 'r-.',label='Third walker')
plt.title("2D Random Walk Starting at (0,0)")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.xlim([0,max(X[:,1])])
plt.show()
```
